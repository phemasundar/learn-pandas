```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("covid_19_data.csv", sep=",")
```

```python
df.head(50)
```

```python
df.drop(["SNo", "Last Update"], axis=1, inplace=True)
```

```python
df.head()
```

```python
df.rename(columns={"ObservationDate": "Date", "Province/State": "State", "Country/Region": "Country"}, inplace=True)
```

```python
df.head()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
df = df.fillna("NA")
df
```

```python
df.info()
```

```python
df2 = df.groupby('Country')[['Confirmed','Deaths','Recovered']].sum()
df2
```

```python
df3 = df.groupby(['Country','Date'])[['Confirmed','Deaths','Recovered']].sum().reset_index()
df3
```

```python
df4 = df2[df2['Confirmed']>50]
df4
```

```python
countries = df3['Country'].unique()
countries
# len(countries)
```

```python
for indCountry in countries:
    allRowsOfCountry = df3[df3['Country'] == indCountry]
    print(allRowsOfCountry)
    plt.scatter(allRowsOfCountry["Date"], allRowsOfCountry["Confirmed"])
    plt.scatter(allRowsOfCountry["Date"], allRowsOfCountry["Deaths"], color = "Red")
    plt.scatter(allRowsOfCountry["Date"], allRowsOfCountry["Recovered"], color="green")
    plt.title(indCountry)
    plt.xlabel("Date")
    plt.ylabel("Number of cases")
    
    # Rotate the x-axis tick labels vertically
    plt.xticks(rotation=45)
    
    plt.legend()
    plt.show()
```
