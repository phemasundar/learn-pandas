```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# array of 1000 values between 0 to 10 in equal intervals
x = np.linspace(0, 10, 1000)
y = np.sin(x)
# plot line chart with x on X-axis & sin(x) on Y-axis
plt.plot(x, y)
```

```python
# plot scatter / dot chart with x on X-axis & sin(x) on Y-axis with eavry 10 element from X and Y arrays
plt.scatter(x[::10], y[::10])
```

```python
# change plot color
plt.scatter(x[::10], y[::10], color = "red")
```

```python
# 2 plots in the same image / step
plt.plot(x, y, color = "blue")
z = np.cos(x)
plt.plot(x, z, color = "green")
```

```python
plt.plot(x, x+0, "-g") # solid green color
plt.plot(x, x+1, "--c") # dashed cyan color
plt.plot(x, x+2, "-.k") # dashdot black color
plt.plot(x, x+3, ":r") # dotted red color
```
